```python
# Milestone 1
# Importing libraries
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data01.csv')
df.head()
```

```python
# dropping NaNs
df.dropna(inplace=True)
df.head()
```

```python
# Creating histogram of age by outcome
age1=df.loc[df.outcome==1.0, 'age']
age0=df.loc[df.outcome==0.0, 'age']
plt.hist(age0,  color='g', label='Alive')
plt.hist(age1,  color='r', label='Not Alive')
plt.gca().set(title='Frequency Histogram of Age', ylabel='Frequency')
plt.legend()
```

```python
# Scatterplot of BMI and outcome
df.plot(kind='scatter', x='BMI',y='outcome',color='red')
plt.title("BMI Vs outcome")
```

```python
# histogram of BMI and diabetes
dia1=df.loc[df.diabetes==1.0, 'BMI']
dia0=df.loc[df.diabetes==0.0, 'BMI']
plt.hist(dia0,  color='b', label='Non-Diabetic')
plt.hist(dia1,  color='g', label='Diabetic')
plt.gca().set(title='Frequency Histogram of BMI', ylabel='Frequency')
plt.legend()
```

```python
plt.title("Outcome by Gender")
sns.countplot(x ='outcome', hue = "gendera", data = df)
plt.show()
```

```python
# Summary
#Frequency histogram of outcome by Age does not provide a conclusive evidence that 
# age was related to the outcome that the person died in the hospital or not.
# BMI vs outcome scatterplot tends to indicate that people with high BMI were also likely to come out alive.Therefore no, conclusive vidence.
# Frequency histogram of BMI again does not show conclusive evidence that BMI causes diabetes; 
# however, it does indicate that people with BMI of ~ 25 to 35 tend to be diabetic.
# Outcome by gender graph does not indicate that gender has any implications on the outcome.
# In summary, individually the selected variables do not have any significant impact on the outcone. 
# Also, BMI and diabetes do not show comorbidity per the graph.
```

```python
#  Changing Column Names to introduce underscope between names with two or more words having a space using lambda function
# This is to avoid using '' between column names having names with two or more words
df.columns = df.columns.map(lambda x : x.replace(" ", "_"))
```

```python
df.describe()
```

```python
# Dropping features which are not indicative of acuteness of the patient's currernt condition and / or are indicative 
# of generic manageable conditions or have features explaining similar characteristics such as:
## group, ID, depression or Gender which does not affect the outcome per the outcome

## Hyperlipemia - Hyperlipidemia is a condition that incorporates various genetic and acquired disorders 
## that describe elevated lipid levels within the human body.

## PT - The prothrombin time, sometimes referred to as PT or pro time, test is a test to evaluate blood clotting. 
## RDW - A red cell distribution width (RDW) test measures the differences in the 
## volume and size of your red blood cells (erythrocytes).

## Similarly Creatine kinase, Urea nitrogen, Blood potassium, Blood sodium, Blood calcium, Chloride, Anion gap, PH,
## Lactic acid, PCO2, and Bicarbonate variables were dropped

df.drop(labels=['group', 'ID', 'gendera', 'depression', 'Hyperlipemia', 'PT', 'RDW', 'Creatine_kinase', 
                'Urea_nitrogen', 'Blood_potassium', 'Blood_sodium', 'Blood_calcium', 'Chloride',
               'Anion_gap', 'PH', 'Lactic_acid','PCO2', 'Bicarbonate'], axis=1, inplace=True)
df.describe()
```

```python
# finding percentage of missing values in each feature
import numpy as np
for column in df.columns:
    print('{} has {} % missing values'.format(column,np.round(df[column].isnull().sum()/len(df)*100,4)))
```

```python
# Drop any features that are missing more than 20% of their values.Only Basophils will drop based on above result.
df=df.drop(df.columns[df.isnull().mean()>0.20],axis=1)
df.describe()
```

```python
df.dtypes
```

```python
# We see that none of the features are object type
```

```python
# dropping duplicate rows if any
df.drop_duplicates(inplace=True)
df.describe()
```

```python
# For numerical columns, fill in any missing data with the median value as the max percentage of missing rows is ~18
# and is not expected to advsersely affect the data
df = df.fillna(df.median(numeric_only=True))
df.describe()
```

```python
# Introducing Urine_output_BMI normalized by BMI as body weight can be a factor in the urine output
df['Urine_output_BMI'] = df['Urine_output']/df['BMI']
df.describe()
```

```python
#  Validating data by Showing descriptive statistics of all columns
df.describe(include = 'all')
```

```python
df.dtypes
```

```python
# Data does not have any missing values. All feature object types are ax expected.Therefore, dataset is ready for modeling.
```

```python
# As noted in Milestone 1, I plan to develop a model which will predict mortalality based on data collected during
# admission of the patient (step 1) and also together with data which is generated typically within first 24 hours based on
# tests done upon admission (step 2).Outcome is a binary variable; therefore, I will be using logisticregression for my model.
```

```python
# Importing the train_test_split Function
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
# step 1
x_1 = df[['age', 'BMI', 'hypertensive', 'atrialfibrillation', 'CHD_with_no_MI', 'diabetes', 'deficiencyanemias', 
        'Renal_failure', 'COPD', 'heart_rate', 'Systolic_blood_pressure', 'Diastolic_blood_pressure', 'Respiratory_rate',
        'temperature', 'SP_O2']]
y_1 = df['outcome']
```

```python
x_1_train, x_1_test, y_1_train, y_1_test = train_test_split(x_1,y_1,test_size=0.2)
```

```python
#we use the standard scaler function to scale the values into a common range.Then we build Model. 
scaler = StandardScaler()
lr = LogisticRegression()
scaler.fit(x_1_train)
x_1_train_scaler = scaler.transform(x_1_train)
x_1_test_scaler = scaler.transform(x_1_test)
```

```python
# Train Logistic Regression on the training data
lr.fit(x_1_train_scaler,y_1_train)
```

```python
#Evaluationg the model using accuracy score to check accuracy and overfitting.
```

```python
x_1_test_pred = lr.predict(x_1_test_scaler)
accuracy = accuracy_score( y_1_test, x_1_test_pred)
accuracy
```

```python
# Accuracy is 88%
```

```python
#Accuracy on model data
```

```python
x_1_train_pred =lr.predict(x_1_train_scaler)
accuracy = accuracy_score( y_1_train, x_1_train_pred)
accuracy
```

```python
#Accuracy is similar to that calculated above. Therefore no overfitting.
```

```python
# Step 2 with all variables
x = df.loc[:,df.columns != 'outcome']
y = df['outcome']
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
lr.fit(x_train_scaler,y_train)
```

```python
x_test_pred = lr.predict(x_test_scaler)
accuracy = accuracy_score( y_test, x_test_pred)
accuracy
```
